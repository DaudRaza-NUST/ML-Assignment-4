```python
import numpy as np 
import random

class NeuralNetwork:

    def __init__(self, X1, X2, y1):
        self.X1 = X1
        self.X2 = X2
        self.y1 = y1
        self.y2 = 1 - y1
        self.W1 = random.uniform(0.1, 0.9)
        self.W2 = random.uniform(0.1, 0.9)
        self.W3 = random.uniform(0.1, 0.9)
        self.W4 = random.uniform(0.1, 0.9)
        self.W5 = random.uniform(0.1, 0.9)
        self.W6 = random.uniform(0.1, 0.9)
        self.W7 = random.uniform(0.1, 0.9)
        self.W8 = random.uniform(0.1, 0.9)
        self.b1 = random.uniform(0.1, 0.9)
        self.b2 = random.uniform(0.1, 0.9)
        self.b3 = random.uniform(0.1, 0.9)
        self.b4 = random.uniform(0.1, 0.9)
        self.epoch = 10000
        self.alpha = 0.075

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))
    
    def forward(self):
        self.z1 = self.W1*self.X1 + self.W2*self.X2 +self.b1
        self.a1 = self.sigmoid(self.z1)
        self.z2 = self.W3*self.X1 + self.W4*self.X2 +self.b2
        self.a2 = self.sigmoid(self.z2)
        self.z3 = self.W5*self.a1 + self.W6*self.a2 +self.b3
        self.o1 = self.sigmoid(self.z3)
        self.z4 = self.W7*self.a1 + self.W8*self.a2 +self.b4
        self.o2 = self.sigmoid(self.z4)
        return self.o1, self.o2
    
    def Mean_Sqaured_Error(self):
        self.MSE_1 = 1/2*(self.y1-self.o1)**2
        self.MSE_2 = 1/2*(self.y2-self.o2)**2
        MSE = (self.MSE_1 + self.MSE_2)
        return MSE
    
    def prime_sigmoid(self, z):
        return self.sigmoid(z)*(1-self.sigmoid(z)) 
    
    def backward(self):
        prime_W8 = (self.o2-self.y2)*self.prime_sigmoid(self.z4)*self.a2
        prime_W7 = (self.o2-self.y2)*self.prime_sigmoid(self.z4)*self.a1
        prime_W6 = (self.o1-self.y1)*self.prime_sigmoid(self.z3)*self.a2
        prime_W5 = (self.o1-self.y1)*self.prime_sigmoid(self.z3)*self.a1
        prime_W4 = (((self.o1-self.y1)*self.prime_sigmoid(self.z3)*self.W6)+((self.o2-self.y2)*self.prime_sigmoid(self.z4)*self.W8))*self.prime_sigmoid(self.z2)*self.X2
        prime_W3 = (((self.o1-self.y1)*self.prime_sigmoid(self.z3)*self.W6)+((self.o2-self.y2)*self.prime_sigmoid(self.z4)*self.W8))*self.prime_sigmoid(self.z2)*self.X1
        prime_W2 = (((self.o1-self.y1)*self.prime_sigmoid(self.z3)*self.W5)+((self.o2-self.y2)*self.prime_sigmoid(self.z4)*self.W7))*self.prime_sigmoid(self.z1)*self.X2
        prime_W1 = (((self.o1-self.y1)*self.prime_sigmoid(self.z3)*self.W5)+((self.o2-self.y2)*self.prime_sigmoid(self.z4)*self.W7))*self.prime_sigmoid(self.z1)*self.X1
        prime_b4 = (self.o2-self.y2)*self.prime_sigmoid(self.z4)
        prime_b3 = (self.o2-self.y2)*self.prime_sigmoid(self.z3)
        prime_b2 = (((self.o1-self.y1)*self.prime_sigmoid(self.z3)*self.W6)+((self.o2-self.y2)*self.prime_sigmoid(self.z4)*self.W8))*self.prime_sigmoid(self.z2)
        prime_b1 = (((self.o1-self.y1)*self.prime_sigmoid(self.z3)*self.W5)+((self.o2-self.y2)*self.prime_sigmoid(self.z4)*self.W7))*self.prime_sigmoid(self.z1)

        self.W1 = self.W1 - (self.alpha*prime_W1)
        self.W2 = self.W2 - (self.alpha*prime_W2)
        self.W3 = self.W3 - (self.alpha*prime_W3)
        self.W4 = self.W4 - (self.alpha*prime_W4)
        self.W5 = self.W5 - (self.alpha*prime_W5)
        self.W6 = self.W6 - (self.alpha*prime_W6)
        self.W7 = self.W7 - (self.alpha*prime_W7)
        self.W8 = self.W8 - (self.alpha*prime_W8)
        self.b1 = self.b1 - (self.alpha*prime_b1)
        self.b2 = self.b2 - (self.alpha*prime_b2)
        self.b3 = self.b3 - (self.alpha*prime_b3)
        self.b4 = self.b4 - (self.alpha*prime_b4)
        return None

    
    def Binary_Classification(self):

        while True:
            Result = self.forward()
            Error = self.Mean_Sqaured_Error()
            self.backward()
            Output1 = Result[0]
            Output2 = Result[1]
            
            if self.epoch == 0 or Error <= 0.000001:
                print("The First output is",Output1)
                print("The Second output is",Output2)
                break
            
            if self.epoch%100 == 0:
                print("Total Error: ",Error)
                print("W1: ", self.W1, " W2: ", self.W2, " W3: ", self.W3, " W4: ", self.W4, " W5: ", self.W5, " W6: ", self.W6, " W7: ", self.W7, " W8: ", self.W8 )
                print("B1: ", self.b1, " B2: ", self.b2, " B3: ", self.b3, " B4: ", self.b4)
            self.epoch -= 1

        return None
    
```

```python
BasicNN = NeuralNetwork(5,15,1)
```
